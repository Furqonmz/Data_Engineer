```python
def luas_keliling(panjang, lebar):
  luas = panjang * lebar
  keliling = 2* (panjang + lebar)
  return luas, keliling
```

```python
luas_keliling(10,11)
```

```python
luas_keliling(10,11)[0]
```

```python
jawaban_l, jawaban_k = luas_keliling(7,4)
print(jawaban_l)
print(jawaban_k)
```
